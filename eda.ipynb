```python
import gc
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
df_train = pd.read_csv('./train.csv')
df_test = pd.read_csv('./test2.csv')
```

```python
df_train['content'] = df_train.content.apply(lambda x: x[:5000])
df_test['content'] = df_test.content.apply(lambda x: x[:5000])
```

```python
sns.distplot(df_train.filename.map(len))
sns.distplot(df_test.filename.map(len))
plt.legend(['train','test'])
plt.show()
```

```python
sns.distplot(df_train.content.map(len))
sns.distplot(df_test.content.map(len))
plt.legend(['train','test'])
plt.show()
```

```python
sns.distplot(df_train.content.apply(lambda s: s.split()).map(len))
sns.distplot(df_test.content.apply(lambda s: s.split()).map(len))
plt.legend(['train','test'])
plt.show()
```
